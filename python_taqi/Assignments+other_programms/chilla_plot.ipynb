# Simple Count Plot with the data of Chilla Zoom meeting

```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns

# Loading the data from the file
chilla= pd.read_csv("Chilla_data.csv")
sns.set_theme(style="ticks", color_codes=True)

sns.countplot(x="Gender", hue="What are you?", data=chilla)
plt.show()

```

# Line Plot using different libraries
### Line plot is always draw between two numerical values

```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns

# Loading the data from the file
chilla= pd.read_csv("Chilla_data.csv")

sns.lineplot(x="Research/Working experience (Float/Int) years", y="How many hours you code a day? ", data= chilla)

# to add the Tile to the Plot

plt.title("Research Exp. to Coding/Day")
plt.show()
```

##### Adding Limits to the the line plot

```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns

# Loading the data from the file
chilla= pd.read_csv("Chilla_data.csv")

sns.lineplot(x="Research/Working experience (Float/Int) years", y="How many hours you code a day? ", data= chilla)

# to add the Tile to the Plot

plt.title("Research Exp. to Coding/Day")
# here setting the x axis limit 
plt.xlim(0.0001)
# here setting the x axis limit
plt.ylim(0.001)
plt.show()
```

#### Setting the style of lineplot
- Darkgrid
- Whitegrid
- White
- Ticks

```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns

# Loading the data from the file
chilla= pd.read_csv("Chilla_data.csv")

sns.lineplot(x="Research/Working experience (Float/Int) years", y="How many hours you code a day? ", data= chilla)
sns.set_style(style=None, rc =None)
sns.set_style("ticks")
# to add the Tile to the Plot

plt.title("Research Exp. to Coding/Day")
plt.show()
```

#### Size of Figure

```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns

# Loading the data from the file
chilla= pd.read_csv("Chilla_data.csv")
# to change the Figure size
plt.figure(figsize=(6,8))

sns.lineplot(x="Research/Working experience (Float/Int) years", y="How many hours you code a day? ", data= chilla)

# to add the Tile to the Plot

plt.title("Research Exp. to Coding/Day")

plt.show()
```

# Box Plot

```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns

# Loading the data from the file
chilla= pd.read_csv("Chilla_data.csv")
sns.set_theme(style="white", color_codes=True)

sns.boxplot(x="Age (years)-Float/Int",y="Where do you live?", hue="Gender", data=chilla)
plt.title( " Age to Gender")
plt.show()
```
